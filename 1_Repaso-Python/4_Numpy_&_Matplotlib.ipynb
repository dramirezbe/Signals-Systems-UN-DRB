```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linspace

lin =np.linspace(0, 1, 10) #np.linspace(start, end, num of points)
print(lin)

ara = np.arange(0, 1, 0.1) #np.arange(start, end, step)
print(ara)
```

```python
#example with fs
fs = 10 # 1000 samples per second
duration = 1 # 1 second
lin = np.linspace(0, duration, int(fs * duration), endpoint=False)
print(lin)

ara = np.arange(0, duration, 1/fs)
print(ara)
```

```python
#nunpy constants
print(np.pi)
print(np.e)
```

```python
#Define functions with linspace and arange
f = 10
w = 2*np.pi*f
fs = 100 * f
signal_duration = 2

#t = np.linspace(-signal_duration/2, signal_duration/2, int(fs * signal_duration), endpoint=False)
t = np.arange(-signal_duration/2, signal_duration/2, 1/fs)

sin_ft = np.sin(w*t)
cos_ft = np.cos(w*t)
plt.plot(t, sin_ft, label='sin')
plt.plot(t, cos_ft, label='cos')
plt.legend()
plt.show()
```

```python
t = np.arange(-1, 1, 1/fs)

dirac = np.zeros_like(t)
dirac[np.abs(t) < 1/fs] = 1  #Nearest value of 0

print(dirac)
plt.plot(t, dirac, label='dirac')
plt.legend()
plt.show()
```

## Function operations

```python
plt.figure(figsize=(20, 6))
plt.plot(t, sin_ft, label='sin')
plt.plot(t, cos_ft, label='cos')
plt.plot(t, step_signal, label='step')
plt.legend()
plt.show()
```

```python
add = sin_ft + step_signal
diff = sin_ft - step_signal

plt.figure(figsize=(18, 6))
plt.plot(t, add, label='add' )
plt.plot(t, diff, label='diff', linestyle='--')
plt.plot(t, step_signal, label='step signal')
plt.plot(t, -step_signal, label='step signal', linestyle='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(t, sin_ft, label='normal sin')
plt.plot(t, 2 * sin_ft, label='Amplitude 2 sin')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(t, sin_ft, label='normal sin')
plt.plot(t, sin_ft + 2, label='Vertical offset sin')
plt.legend()
plt.show()
```

```python
step_horizontal_offset = np.heaviside(t - 0.25, 1)
plt.figure(figsize=(15, 6))
plt.plot(t, step_horizontal_offset, label='horizontal offset step', linestyle='--')
plt.plot(t, step_signal, label='normal step')
plt.legend()
plt.show()
```

```python
#Portioning signal

plt.figure(figsize=(15, 6))
#plt.plot(t, sin_ft, label='normal sin')
t_half = t[:len(t)//2]
sin_ft_half = sin_ft[:len(t)//2]
t_other_half = t[len(t)//2:]
sin_ft_other_half = sin_ft[len(t)//2:]
plt.plot(t_half, sin_ft_half, label='first half sin', linestyle='--')
plt.plot(t_other_half, sin_ft_other_half, label='second half sin', linestyle='--')
plt.legend()
plt.show()
```
